### Outlier Detection and handling

```python
sns.boxplot(data["Absences"])
```

#### *ManagerID 

```python
sns.boxplot(data["ManagerID"])
```

```python
Q1=np.percentile(data["ManagerID"],25,interpolation="midpoint")
Q2=np.percentile(data["ManagerID"],50,interpolation="midpoint")
Q3=np.percentile(data["ManagerID"],75,interpolation="midpoint")
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
outlier=[]
for x in data["ManagerID"]:
    if(x>up_lim)or(x<low_lim):
        outlier.append(x) 

index2 = (data["ManagerID"]>up_lim)
index2 = data.loc[index2].index
data.loc[list(index2),'ManagerID']=up_lim

sns.boxplot(data['ManagerID'])
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Employee_Name"]=le.fit_transform(data["Employee_Name"])
data["Position"]=le.fit_transform(data["Position"])
data["State"]=le.fit_transform(data["State"])
data["DOB"]=le.fit_transform(data["DOB"])
data["Sex"]=le.fit_transform(data["Sex"])
data["MaritalDesc"]=le.fit_transform(data["MaritalDesc"])
data["HispanicLatino"]=le.fit_transform(data["HispanicLatino"])
data["DateofHire"]=le.fit_transform(data["DateofHire"])
data["TermReason"]=le.fit_transform(data["TermReason"])
data["Department"]=le.fit_transform(data["Department"])
data["RecruitmentSource"]=le.fit_transform(data["RecruitmentSource"])
data["ManagerName"]=le.fit_transform(data["ManagerName"])
data["PerformanceScore"]=le.fit_transform(data["PerformanceScore"])
data["LastPerformanceReview_Date"]=le.fit_transform(data["LastPerformanceReview_Date"])
data["CitizenDesc"]=le.fit_transform(data["CitizenDesc"])
data["RaceDesc"]=le.fit_transform(data["RaceDesc"])
data["DateofTermination"]=le.fit_transform(data["DateofTermination"])
data["EmploymentStatus"]=le.fit_transform(data["EmploymentStatus"])
```

```python
x=data.drop("Salary",axis=1)
y=data["Salary"]
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
print('Training Accuracy: ', lin_reg.score(x_train, y_train))
print('Testing Accuracy: ', lin_reg.score(x_test, y_test))
```

```python
from sklearn.linear_model import RidgeCV, Ridge
ridge= Ridge(alpha =0.001,  normalize = True).fit(x_train,y_train)
print('Training Accuracy: ', ridge.score(x_train, y_train))
print('Testing Accuracy: ', ridge.score(x_test, y_test))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha = 0.01).fit(x_train, y_train)
print('Training Accuracy: ', lasso.score(x_train, y_train))
print('Testing Accuracy: ', lasso.score(x_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state = 42)
rf.fit(x_train, y_train)
print('Training Accuracy:', rf.score(x_train, y_train))
print('Testing Accuracy: ', rf.score(x_test, y_test))
```

```python
acc_table = pd.DataFrame({'Model': ['Linear','Ridge','Lasso','Random Forest',],
                         'Accuracy Score': [lin_reg.score(x_train, y_train),ridge.score(x_train, y_train),lasso.score(x_train, y_train),rf.score(x_train, y_train)]})
acc_table = acc_table.sort_values(by='Accuracy Score', ascending=False)
acc_table.style.background_gradient(cmap='Blues')
```

```python
# Fine tuning the hyper parameters of the model
rf1=RandomForestRegressor(n_estimators = 1000, oob_score = True,n_jobs = 1,random_state =1)  
rf1.fit(x_train,y_train)
y_pred=rf1.predict(x_test)
print('Training Accuracy: ', rf1.score(x_train, y_train))
print('Testing Accuracy: ', rf1.score(x_test, y_test))
```
